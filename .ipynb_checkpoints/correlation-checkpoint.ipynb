```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
currency = pd.read_csv('currency_parsed.csv')
currency.describe()
```

```python
id = currency[['id']].values.astype(int)
usd = currency[['Usd']].values.astype(float)
eur = currency[['Eur']].values.astype(float)
cny = currency[['Cny']].values.astype(float)
krw = currency[['Krw']].values.astype(float)
brl = currency[['Brl']].values.astype(float)

min_max_scaler = preprocessing.MinMaxScaler()

id = id.flatten()
usd = min_max_scaler.fit_transform(usd).flatten()
eur = min_max_scaler.fit_transform(eur).flatten()
cny = min_max_scaler.fit_transform(cny).flatten()
krw = min_max_scaler.fit_transform(krw).flatten()
brl = min_max_scaler.fit_transform(brl).flatten()

data = {'id':id, 'Usd':usd, 'Eur':eur, 'Cny':cny, 'Krw':krw, 'Brl':brl}

currency_normalized = pd.DataFrame(data, index=id)
currency_normalized.describe()
```

```python
wowtoken = pd.read_csv('wowtoken_parsed.csv')
wowtoken.describe()
```

```python
id = wowtoken[['id']].values.astype(int)
us = wowtoken[['Us']].values.astype(float)
eu = wowtoken[['Eu']].values.astype(float)
ch = wowtoken[['Ch']].values.astype(float)
kr = wowtoken[['Kr']].values.astype(float)
tw = wowtoken[['Tw']].values.astype(float)

min_max_scaler = preprocessing.MinMaxScaler()

id = id.flatten()
us = min_max_scaler.fit_transform(us).flatten()
eu = min_max_scaler.fit_transform(eu).flatten()
ch = min_max_scaler.fit_transform(ch).flatten()
kr = min_max_scaler.fit_transform(kr).flatten()
tw = min_max_scaler.fit_transform(tw).flatten()

data = {'id':id, 'Us':us, 'Eu':eu, 'Ch':ch, 'Kr':kr, 'Tw':tw}

wowtoken_normalized = pd.DataFrame(data, index=id)
wowtoken_normalized.describe()
```

```python
s1 = pd.Series(currency_normalized["Eur"], index=currency_normalized["id"])
print(s1)
```

```python
s2 = pd.Series(wowtoken_normalized["Eu"], index=wowtoken_normalized["id"])
print(s2)
```

```python
print("pearson = %f" % s1.corr(s2, method="pearson"))
print("kendall = %f" % s1.corr(s2, method="kendall"))
print("spearman = %f" % s1.corr(s2, method="spearman"))
```
